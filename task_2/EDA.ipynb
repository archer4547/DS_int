```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("alessiocorrado99/animals10")

print("Path to dataset files:", path)
```

```python
import os
from PIL import Image
```

```python
os.listdir(path)
```

```python

x_sizes = set()
y_sizes = set()

all_classes_path = os.path.join(path, 'raw-img')

for class_name in os.listdir(all_classes_path):
    class_path = os.path.join(all_classes_path, class_name)
    
    for filename in os.listdir(class_path):
        img_path = os.path.join(class_path, filename)
        try:
            img = Image.open(img_path)
            x, y = img.size
            x_sizes.add(x)
            y_sizes.add(y)
            # sizes.add(img.size)
        except Exception as e:
            print(f"Skipping {filename}: {e}")
                    
print('Unique x sizes count:', len(x_sizes))
print('Unique y sizes count:', len(y_sizes))

print('Minimum x size:', min(x_sizes))
print('Minimum y size:', min(y_sizes))
```
