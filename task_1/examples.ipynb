```python
# External libs
from tensorflow.keras import datasets
```

```python
# Internal lib
from MnistClassifiers import MnistClassifier
```

```python
# Loading dataset
(x_train, y_train), (x_test, y_test) = datasets.mnist.load_data()
```

```python
# Normalization
x_train = x_train.astype('float64') / 255
x_test = x_test.astype('float64') / 255
```

```python
# Correct classifiers
cnn = MnistClassifier('cnn')
rf = MnistClassifier('rf')
nn = MnistClassifier('nn')
```

```python
# Training the model
nn.train(x_train, y_train)
```

```python
# Predictions using the model
for i in range(5):
    print(f'real {y_test[i]} | predicted {nn.predict(x_test[i : i+1])}')
```

```python
cnn.train(x_train, y_train)
```

```python
for i in range(5):
    print(f'real {y_test[i]} | predicted {cnn.predict(x_test[i : i+1])}')
```

```python
# flattening the X for random forest
x_train_flat = x_train.reshape(x_train.shape[0], -1)
x_test_flat = x_test.reshape(x_test.shape[0], -1)
```

```python
rf.train(x_train_flat, y_train)
print('Test accuracy:', rf.model.model.score(x_test_flat, y_test))
```

```python
for i in range(5):
    print(f'real {y_test[i]} | predicted {rf.predict(x_test_flat[i : i+1])}')
```
